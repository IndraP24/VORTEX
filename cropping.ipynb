```python
import cv2
import numpy as np
import os
import matplotlib.pylab as plt
```

```python
def crop(img,x1,y1,x2,y2):
    return img[y1:y2,x1:x2]
```

```python
target_dir = "./data/processed/crop_data/"
source_dir = "./data/raw/final_data/"
```

```python
for x in os.listdir(source_dir):
    print(x)
```

```python
for x in os.listdir(source_dir):
    print(x)
    for y in os.listdir(source_dir+x):
        #print(y)
        img = cv2.imread(source_dir+x+'/'+y,cv2.IMREAD_GRAYSCALE)
        img_new = crop(img,220,250,465,520)
        cv2.imwrite(target_dir+x+'/'+y,img_new)

```

```python
def display_img(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
#img_gray[250:520,220:465]
im = cv2.imread('data/raw/final_data/EXTREME/3DIMG_04NOV2019_0000_L1B_STD_IR2.jpg',cv2.IMREAD_GRAYSCALE)
    
```

```python
display_img(im)
```

```python
im_rec = cv2.rectangle(im.copy(),(220,250),(465,520),(0,255,0),3)
```

```python
display_img(im_rec)
```

```python
resized_im = crop(im,220,250,465,520)
```

```python
display_img(resized_im)
```

```python
ret, thresh1 = cv2.threshold(resized_im,127,255,cv2.THRESH_TOZERO)
display_img(thresh1)
```

```python
# gamma 
g = 2.2
g_img = np.power(resized_im,g)
```

```python
display_img(g_img)
```

```python
def applyTrheshZero(img):
    ret, n_img = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
    return n_img
```

```python
target_dir = "./data/processed/threshZero_data/"
source_dir = "./data/processed/crop_data/"
```

```python
# Applying thresh to zero thresholding to all images
for x in os.listdir(source_dir):
    print(x)
    
    for y in os.listdir(source_dir+x):
        #print(y)
        img = cv2.imread(source_dir+x+'/'+y,cv2.IMREAD_GRAYSCALE)
        img_new = applyTrheshZero(img)
        cv2.imwrite(target_dir+x+'/'+y,img_new)
```

```python
img_thres = applyTrheshZero(resized_im)
display_img(img_thres)
```
